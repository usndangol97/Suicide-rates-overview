```python
total_suicide_till_date = {}

for country in df.country.unique():
    suicide_count = 0
    for i in range(len(df[df["country"]==country])):
        suicide_count += df["suicides_no"][i]
        
    total_suicide_till_date[country] = suicide_count
    
df_suicide_till_date = pd.DataFrame.from_dict(total_suicide_till_date ,orient="index")
df_suicide_till_date.columns = ["total_suicide_till_date"]
df_suicide_till_date.sort_values(by="total_suicide_till_date", ascending=False, inplace=True)
df_suicide_till_date.tail(20)
```

```python
df_suicide_till_date.iloc[:10].plot(kind="bar")
plt.title("Top 10 countries with highesst suicide_no till date")
plt.xlabel("suicide_no")
plt.ylabel("Countries")
```

```python
df_suicide_till_date.iloc[91:].plot(kind="bar")
plt.title("Top 10 countries with Least suicide_no till date")
plt.xlabel("suicide_no")
plt.ylabel("Countries")
```
