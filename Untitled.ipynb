```python
df_2008 = df[df["year"]==2008]
print("Lowest GDP per capita in year 2008: ",df_2008["gdp_per_capita ($)"].min())
print("Highest GDP per capita in year 2008: ",df_2008["gdp_per_capita ($)"].max())
```

```python
df_low_gdp_2008 = df_2008[df_2008["gdp_per_capita ($)"]==1096]
df_high_gdp_2008 = df_2008[df_2008["gdp_per_capita ($)"]==121315]

df_high_low_gdp = pd.concat([df_high_gdp_2008, df_low_gdp_2008])

sns.relplot(y="suicides_no", x="gdp_per_capita ($)", kind="scatter", data=df_high_low_gdp, ci=None, hue="country")
plt.title("sucides rates in country with highest vs lowest GDP in year 2008")
```
